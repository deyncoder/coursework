## Extract Data Using Web Scraping

The wikipedia webpage https://en.wikipedia.org/wiki/List_of_largest_banks provides information about largest banks in the world by various parameters. Scrape the data from the table 'By market capitalization' and store it in a JSON file.

Steps:
1. Define url
2. Get the contents of the webpage in text format and store in a variable called data
3. Create soup object
4. Find all html tables in the web page, store to variable tables. In html, table is represented by the tag \<table>.
5. Get length of table list
6. Determine table index of target table
7. Display target table given index and prettify
8. Make dataframe of the table
9. Save dataframe to json


```python
url = "https://en.wikipedia.org/wiki/List_of_largest_banks"
```

```python
html_data = requests.get(url).text
html_data[101:124]
```

```python
soup = BeautifulSoup(html_data, "html.parser")
```

```python
tables = soup.find_all('table')
tables
```

#### len(tables)

```python
for index,table in enumerate(tables):
    if ("Market cap" in str(table)):
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
bankmarketcap_data = pd.DataFrame(columns=["Rank", "Bank Name", "Market Cap (US $ Billion)"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        rank = col[0].text.strip()
        bankname = col[1].text.strip()
        marketcap = col[2].text.replace("[6]","")
        bankmarketcap_data = bankmarketcap_data.append({"Rank":rank, "Bank Name":bankname, "Market Cap (US $ Billion)":marketcap}, ignore_index=True)

bankmarketcap_data.style.hide_index()
```

```python
# load dataframe to json

bankmarketcap_data.to_json("bank_market_cap.json", orient="table")
```
